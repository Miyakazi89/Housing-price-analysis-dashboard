```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data= pd.read_csv('Housing.csv')

data.head()
```

```python
data.info()
data.describe()
```

```python
data.isnull().sum()
data.duplicated()
```

#  Univariate Analysis

```python
sns.histplot(data['price'], kde=True)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data['area'], kde=True)
plt.title("Area Distribution")
plt.show()
```

```python
sns.countplot(x='bedrooms', data=data)
plt.title("Number of Bedrooms")
plt.show()
```

```python
sns.countplot(x='bathrooms', data=data)
plt.title("Bathrooms Distribution")
plt.show()
```

```python
sns.countplot(x='stories', data=data)
plt.title("Stories Distribution")
plt.show()
```

```python
sns.countplot(x='parking', data=data)
plt.title("Parking Spaces")
plt.show()
```

```python
sns.boxplot(y=data['price'])
plt.title("Price Outliers")
plt.show()
```

```python
sns.boxplot(y=data['area'])
plt.title("Area Outliers")
plt.show()
```

# Bivariate Analysis


```python
sns.scatterplot(x='area', y='price', data=data)
```

The scatter plot reveals a positive relationship between property area and price, indicating that larger houses tend to be more expensive. As the area increases, the price generally rises, although there is some variation due to other influencing factors such as location, number of rooms, and amenities. Additionally, a few outliers can be observed, representing unusually expensive or large properties.

```python

sns.barplot(x='bedrooms', y='price', data=data)
plt.title("Average Price by Number of Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Average Price")
plt.show()
```

```python
sns.boxplot(x='bedrooms', y='price', data=data)
plt.title("Price Distribution by Bedrooms")
plt.show()
```

```python
sns.barplot(x='bathrooms', y='price', data=data)
plt.title("Average Price by Number of Bathrooms")
plt.xlabel("Bathrooms")
plt.ylabel("Average Price")
plt.show()
```

```python
sns.boxplot(x='bathrooms', y='price', data=data)
plt.title("Price Distribution by Bathrooms")
plt.show()
```
